# #The next is the data image preprocessing

```python
import torch
import torchvision
from torchvision import datasets, models, transforms
import torch.nn as nn
torch.__version__
```

```python
train_on_gpu = torch.cuda.is_available()

if not train_on_gpu:
    print('CUDA is not available.  Training on CPU ...')
else:
    print('CUDA is available!  Training on GPU ...')
```

```python
classes = os.listdir(VALID_DATA_DIR)
```

```python
data_transform = transforms.Compose([transforms.RandomResizedCrop(224),
                                     transforms.ToTensor(),
                                     transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                     std=[0.229, 0.224, 0.225])])

train_data = datasets.ImageFolder(TRAIN_DATA_DIR, transform=data_transform)
test_data = datasets.ImageFolder(VALID_DATA_DIR, transform=data_transform)

# print out some data stattistics to check whether is model is working
print('Num training images: ', len(train_data))
print('Num test images: ', len(test_data))
```

```python
batch_size = 32
num_workers=0

# prepare data loaders easy and hopefull
train_loader = torch.utils.data.DataLoader(train_data, batch_size=batch_size, 
                                           num_workers=num_workers, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_data, batch_size=batch_size, 
                                          num_workers=num_workers, shuffle=True)
```

```python
dataiter = iter(train_loader)
images, labels = dataiter.next()
images = images.numpy() # convert images to the numerical python for the display

fig = plt.figure(figsize=(25, 4))
for idx in np.arange(20):
    ax = fig.add_subplot(2, 20/2, idx+1, xticks=[], yticks=[])
    plt.imshow(np.transpose(images[idx], (1, 2, 0)))
    ax.set_title(classes[labels[idx]])
```

```python
vgg16 = models.vgg16(pretrained=True)
print(vgg16)
```

```python
print(vgg16.classifier[6].in_features) 
print(vgg16.classifier[6].out_features)
```

```python
for param in vgg16.features.parameters():
    param.requires_grad = False
```

```python
n_inputs = vgg16.classifier[6].in_features
last_layer = nn.Linear(n_inputs, len(classes))

# i am using the model which is VGG16 u can try yours no problem to see the difference
vgg16.classifier[6] = last_layer

# if GPU is available, move the model to GPU
if train_on_gpu:
    vgg16.cuda()

print(vgg16.classifier[6].out_features)
```

```python
import torch.optim as optim
criterion = nn.CrossEntropyLoss()

# here we need to specify optimizer (stochastic gradient descent) and learning rate = 0.01 or 0.001
optimizer = optim.SGD(filter(lambda p: p.requires_grad,vgg16.parameters()), lr = 0.01, momentum=0.001)
```

```python
epochs = 20
valid_loss_min = np.Inf 
```

```python
epoch_list = []
train_loss_list = []
val_loss_list = []
train_acc_list = []
val_acc_list = []
```

```python
for epoch in range(epochs):
    train_loss = 0.0
    val_loss = 0.0
    vgg16.train()        
    correct = 0
    total = 0
```

```python
for data, target in train_loader:   
        if train_on_gpu:
            data, target = data.cuda(), target.cuda()
        optimizer.zero_grad()
        output = vgg16(data)
        if type(output) == tuple:
            output, _ = output
```

```python
       predicted = torch.max(output.data, 1)[1]        
        # Total number of labels
        total += len(target)
        # Total correct predictions
        correct += (predicted == target).sum()
        
        # calculate the loss
        loss = criterion(output, target)
       
        loss.backward()
  
        optimizer.step() 
        train_loss += loss.item()*data.size(0)
    train_loss = train_loss/len(train_loader.dataset)
    
    
    accuracy = 100 * correct / float(total)
    train_acc_list.append(accuracy)
    train_loss_list.append(train_loss)   
    vgg16.eval()         
    correct = 0
    total = 0
    with torch.no_grad():
        for data, target in test_loader:
            if train_on_gpu:
                data, target = data.cuda(), target.cuda()
            # Predict Output
            output = vgg16(data)
            if type(output) == tuple:
                output, _ = output
            # Calculate Loss
            loss = criterion(output, target)
            val_loss += loss.item()*data.size(0)
            # Get predictions from the maximum value
            predicted = torch.max(output.data, 1)[1]

            # Total number of labels
            total += len(target)

            # Total correct predictions
            correct += (predicted == target).sum()
    
    # calculate average training loss and accuracy over an epoch
    val_loss = val_loss/len(test_loader.dataset)
    accuracy = 100 * correct/ float(total)
    
    # Put them in their list
    val_acc_list.append(accuracy)
    val_loss_list.append(val_loss)
    
    # Print the Epoch and Training Loss Details with Validation Accuracy   
    print('Epoch: {} \tTraining Loss: {:.4f}\t Val. acc: {:.2f}%'.format(
        epoch+1, 
        train_loss,
        accuracy
        ))
    # save model if validation loss has decreased
    if val_loss <= valid_loss_min:
        print('Validation loss decreased ({:.6f} --> {:.6f}).  Saving model ...'.format(
        valid_loss_min,
        val_loss))
        # Save Model State on Checkpoint
        
```

```python
torch.save(vgg16.state_dict(), 'file.pt')
valid_loss_min = val_loss
    # Move to next epoch
epoch_list.append(epoch + 1)
```

```python
vgg16.load_state_dict(torch.load('file.pt'))
```

```python
torch.save(vgg16, 'classifier.pth')
```

```python
plt.plot(epoch_list,train_loss_list)
plt.plot(val_loss_list)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training/Validation Loss vs Number of Epochs")
plt.legend(['Train', 'Valid'], loc='upper right')
plt.show()
```

```python
plt.plot(epoch_list,train_acc_list)
plt.plot(val_acc_list)
plt.xlabel("Epochs")
plt.ylabel("Training/Validation Accuracy")
plt.title("Accuracy vs Number of Epochs")
plt.legend(['Train', 'Valid'], loc='best')
plt.show()
```

```python
val_acc = sum(val_acc_list[:]).item()/len(val_acc_list)
print("Validation Accuracy of model = {} %".format(val_acc))
```

```python
dataiter = iter(test_loader)
images, labels = dataiter.next()
img = images.numpy()

if train_on_gpu:
    images = images.cuda()

vgg16.eval()
output = vgg16(images)
if type(output) == tuple:
            output, _ = output
_, preds_tensor = torch.max(output, 1)
preds = np.squeeze(preds_tensor.numpy()) if not train_on_gpu else np.squeeze(preds_tensor.cpu().numpy())

fig = plt.figure(figsize=(20, 5))
for idx in np.arange(12):
    ax = fig.add_subplot(3, 4, idx+1, xticks=[], yticks=[])
    plt.imshow(np.transpose(img[idx], (1, 2, 0)))
    ax.set_title("Pr: {} Ac: {}".format(classes[preds[idx]], classes[labels[idx]]),
                 color=("green" if preds[idx]==labels[idx].item() else "red"))
```
